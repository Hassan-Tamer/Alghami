```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df30 = pd.read_csv('Dataset/features_30_sec.csv')
df3 = pd.read_csv('Dataset/features_3_sec.csv')
df30.head()
```

```python
df30.describe()
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df30['label']=le.fit_transform(df30['label'])
```

```python
df30.drop('filename',axis=1,inplace=True)
```

```python
df30.drop(['rms_var' , "zero_crossing_rate_var" , "perceptr_mean" , "perceptr_var"], axis=1, inplace=True)
```

```python
X = df30.drop('label',axis=1).values
y = df30['label'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state=0)
```

```python
# Normalizing all the data (mean approx 0, std approx 1)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train[:1])
print("Shape = ",X_train.shape)
print("Mean = ",X_train.mean().round(2))
print("STD = ",X_train.std())
```

```python
from sklearn.decomposition import PCA
n_components = 12
pca = PCA(n_components=n_components)
X_train_PCA = pca.fit_transform(X_train)
X_test_PCA = pca.transform(X_test)
```
